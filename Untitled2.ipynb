```python
import os
import sys
import numpy as np
from scipy.interpolate import bisplrep, bisplev, interp2d
import tables
import h5py

import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from matplotlib.ticker import LogFormatterMathtext
plt.style.use('paper.mplstyle')

import nuSQUIDSpy as nsq
from physicsconstants import PhysicsConstants
```

```python
units = PhysicsConstants()

nu_flux = np.genfromtxt('data/PostPropagation/SIBYLL2.3_pp_CombinedGHAndHG_H4a_nu.txt')
nu_bar_flux = np.genfromtxt('data/PostPropagation/SIBYLL2.3_pp_CombinedGHAndHG_H4a_nubar.txt')
flux = np.hstack([nu_flux, nu_bar_flux[:,1:]])
# Get inital flux into the form nuSQuIDS expects
# initial_flux = np.ones((len(flux[:,0]), 2, 3))
initial_flux = np.zeros((len(flux[:,0]), 2, 3))
for i in range(3):
    initial_flux[:,0,i] = flux[:, i+1]
    initial_flux[:,1,i] = flux[:, i+4]
```

```python
interactions = True

# Best fit values from http://www.nu-fit.org/?q=node/211 for NO without SK atmospheric data
delta_m_21 = 7.39e-5
delta_m_31 = 2.523e-3
theta_12   = np.radians(33.82)
theta_13   = np.radians(8.61)
theta_23   = np.radians(48.3)
delta_cp   = np.radians(222)

zens = np.radians(np.linspace(80, 180, 41))
r_e    = 6371*units.km
```

```python
ee = np.logspace(np.log10(10177900000/units.GeV),5,50)

numuflux = np.zeros((2, len(zens), len(flux[:,0])))

for i,zen in enumerate(zens):
    chord  = 2*r_e*np.sin(zen/2)
    nuSQ = nsq.nuSQUIDS(nu_flux[:,0]*units.GeV, 3, nsq.NeutrinoType.both, interactions)
    nuSQ.Set_abs_error(1.0e-10)
    nuSQ.Set_rel_error(1.0e-10)

    nuSQ.Set_SquareMassDifference(1, delta_m_21)
    nuSQ.Set_SquareMassDifference(2, delta_m_31)
    nuSQ.Set_MixingAngle(0, 1, theta_12)
    nuSQ.Set_MixingAngle(0, 2, theta_13)
    nuSQ.Set_MixingAngle(1, 2, theta_23)
    nuSQ.Set_CPPhase(0,2,delta_cp)

    nuSQ.Set_Body(nsq.Earth())
    nuSQ.Set_Track(nsq.Earth.Track(chord))
    nuSQ.Set_TauRegeneration(True)
    nuSQ.Set_initial_state(initial_flux, nsq.Basis.flavor)
    nuSQ.EvolveState()
    
    numuflux[0,i,:] = [nuSQ.EvalFlavor(1,e*units.GeV,0) for e in flux[:,0]]
    numuflux[1,i,:] = [nuSQ.EvalFlavor(1,e*units.GeV,1) for e in flux[:,0]]
```

```python
x, y     = np.mgrid[zens[0]:zens[-1]:41j, np.log(ee)[0]:np.log(ee)[-1]:50j]
z        = np.log(numuflux)

bitck_nu    = bisplrep(x, y, z[0], s=0)
bitck_nubar = bisplrep(x, y, z[1], s=0)

def get_flux(zen, e, pid):
    if e<1e1 or e>1e5:
        return 0
    else:
        if pid==14:
            bitck = bitck_nu
        elif pid==-14:
            bitck = bitck_nubar
        else:
            print('wrong pid')
            quit()
        return np.exp(bisplev(zen, np.log(e), bitck))
```
