```python

import matplotlib
import numpy
import os
import pylab
%matplotlib inline
```

```python
Files=os.listdir("../")
Nus=[r"$\nu_e$",r"$\nu_\mu$",r"$\nu_\tau$"]
FNames=['e','mu','tau']
for inu in range(0,3):
    pylab.figure(figsize=(10,5))
    for f in Files:
        if("_nu.txt" in f):
            dat=numpy.loadtxt("../"+f)
            pylab.loglog(dat[:,0],dat[:,0]*dat[:,0]*dat[:,0]*dat[:,inu+1]*1e-4,label=f) 
            pylab.ylim(1e-7, 1e-5)
            pylab.xlim(10,1e10)
            pylab.title(Nus[inu],fontsize=28)
    pylab.xlabel(r"$E_\nu$ / GeV",fontsize=16)
    pylab.ylabel(r"$\Phi_{\nu}$ ($E_\nu$/GeV)$^3$ (cm$^{2}$ s sr GeV)$^{-1}$",fontsize=16)
    pylab.legend(loc='right',fontsize=11.5)
    pylab.grid(which='both')
    pylab.tight_layout()
    pylab.savefig("SummaryPlot_nu_"+FNames[inu]+".pdf")
    pylab.savefig("SummaryPlot_nu_"+FNames[inu]+".png")
    pylab.show()
```
