```python
ahlers_175 = np.loadtxt('/data/user/isafa/gzk_digitized/ahlers2010/ahlers2010_emin_17.5.csv', delimiter=',')
ahlers_175 = np.delete(ahlers_175, -8, 0)
ahlers_175_log_e = np.log10(np.array(list(zip(*ahlers_175))[0])[::])
ahlers_175_log_f = np.log10(np.array(list(zip(*ahlers_175))[1])[::])
log_ahlers_175_spline = UnivariateSpline(ahlers_175_log_e, ahlers_175_log_f, k = 4, s=0.001)

min_log_e = np.min(ahlers_175_log_e)
max_log_e = np.max(ahlers_175_log_e)
test_log_ens = np.linspace(min_log_e, max_log_e, 500)

e_phi = np.power(10., log_ahlers_175_spline(test_log_ens)) / np.power(10, test_log_ens) / np.power(10, test_log_ens)
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

plt.plot(np.power(10, test_log_ens), e_phi, label = 'Ahlers 17.5')

h = plt.hist(eini / 1e9, bins = ebins, histtype='step', linewidth=3., weights = 1e9/eini)
g = plt.hist(eout / 1e9, bins = ebins, histtype='step', linewidth=3., weights = 1e9/eout)

plt.xlabel(r'$E_{\nu}$ (GeV)')
plt.ylabel(r'$E_{\nu} d\Phi / dE_{\nu}$ (GeV cm$^{-2}$ s$^{-1}$ sr$^{-1}$) $\times$ C')

plt.loglog()
plt.show()

normalization = np.max(e_phi) / (np.max(h[0])/np.log(10))
print(normalization)
```
