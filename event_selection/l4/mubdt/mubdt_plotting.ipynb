```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('data_final.csv')
data.head(15)
```

```python
data.drop(columns=data.columns[0], inplace=True)
data.head(15)
```

```python
train_index = np.load('train_index.npy')
test_index = np.load('test_index.npy')
```

```python
data_test = data.iloc[test_index]
```

```python
test_sig = data_test.loc[data_test['muon'] == 1]
test_bkg = data_test.loc[data_test['muon'] == 0]
```

```python
plt.hist(test_sig['muon_prob'], weights=test_sig['norm-rate'], label='Muon samples', bins=300, histtype='step', density=True, cumulative=True)
plt.hist(test_bkg['muon_prob'], weights=test_bkg['norm-rate'], label='Neutrino samples', bins=300, histtype='step', density=True, cumulative=True)
plt.yscale('symlog')
plt.yticks(np.logspace(-1, 0, 10), 
           ['$10^{-1}$', '', '', '', '', '', '', '', '', '$10^0$'])

plt.xlabel('$\mu$BDT predicted probability event is muon-like')
plt.ylabel('Cumulative probability')
plt.legend(ncol=2, loc='upper center', bbox_to_anchor=(0.5, -0.17))
plt.show()
```
